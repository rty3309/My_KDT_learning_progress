# 정규 표현식

##### 정규 표현식 예제 1

```python
# 정규 표현식 라이브러리 불러오기
import re
```

```python
# compile() 사용 안함
m = re.match('[a-z]+', 'Python')
print(m)    # ‘Python’은 대문자로 시작하기 때문에 match()함수의 리턴값이 None
print(re.search('apple', 'I like apple!'))    # span=( , ) : 패턴과 일치하는 문자열의 범위

# compile() 사용 : 객체 생성
p = re.compile('[a-z]+')    # 알파벳 소문자,  + : 패턴
m = p.match('python')    # 패턴과 비교할 문자열
print(m)
print(p.search('I like apple 123'))
```

##### 정규 표현식 예제 2

```python
m = re.match('[a-z]+', 'pythoN')    # 'pythoN'의 처음부터 소문자 검사
print(m)
```

```python
m = re.match('[a-z]+', 'pYthon')    # 첫 문자는 소문자
print(m)
```

```python
print(re.match('[a-z]+', 'regex python'))
print(re.match('[a-z]+', ' regexpython'))

print(re.match('[a-z]+', 'regexpythoN'))
print(re.match('[a-z]+$', 'regexpythoN'))    # $ : 문자열의 마지막에 소문자 1회 이상 검사

print(re.match('[a-z]+', 'regexPython'))
print(re.match('[a-z]+$', 'regexpython'))
```

##### 정규 표현식 예제 3

```python
p = re.compile('[a-z]+')    # 알파벳 소문자
print(p.findall('life is too short!. Regular expression test'))

find_all = re.findall('[a-z]+', 'life is too short!. Regular expression test')
# 이거는 패턴 생성 안 한 예제. findall 내부에 집어넣음. 결과는 같음
print(find_all)
```

```python
result = p.search('I like apple 123')
print(result)

result = p.findall('I like apple 123')
print(result)
```

##### Match 메소드 예제 1

```python
# ^ .. $ 을 명시해야 정확한 자리수 검사가 이루어짐
tel_checker = re.compile(r'^(\d{2,3})-(\d{3,4})-(\d{4})$')
# tel_checher : 객체, r : raw string

print(tel_checker.match('02-123-4567'))
print(tel_checker.match('02-123-4567').group())    # 바로 쓰는 것은 group()이 제일 편함
print(tel_checker.match('02-123-4567').groups())

print(tel_checker.match('053-950-45678'))
print(tel_checker.match('053950-4567'))
```

##### Match 메소드 예제 2

```python
tel_number = '053-950-4567'
tel_number = tel_number.replace('-', '')
print(tel_number)

tel_checker1 = re.compile(r'^(\d{2,3})(\d{3,4})(\d{4})$')
print(tel_checker1.match(tel_number))
print(tel_checker1.match('0239501234'))    # 02-3950-1234
print(tel_checker1.match('0239501234').group(0))
```

##### Match 메소드 예제 3

```python
tel_checker = re.compile(r'^(\d{2,3})-(\d{3,4})-(\d{4})$')
m = tel_checker.match('02-123-4567')

print(m.groups())
print('group(): ', m.group())    # group(0)과 동일 결과
print('group(0): ', m.group(0))
print('group(1): ', m.group(1))
print('group(2,3): ', m.group(2,3))
print('start(): ', m.start())    # 매칭된 문자열의 시작 인덱스
print('end(): ', m.end())    # 매칭된 문자열의 마지막 인덱스+1

                #      [0]         [1]    [2]    [3]
# group 내부 구조 ('02-123-4567', '02', '123', '4567')
```

```python
cell_phone = re.compile(r'^(01(?:0|1|[6-9]))-(\d{3,4})-(\d{4})$')

print(cell_phone.match('010-123-4567'))
print(cell_phone.match('019-1234-5678'))
print(cell_phone.match('001-123-4567'))
print(cell_phone.match('010-1234567'))
```

##### 전방 탐색(lookahead)

```python
# 전방 긍정 탐색(?=)
lookahead1 = re.search(r'.+(?=won)', '1000won') # .+(?=won) : 패턴('won'문자열 앞에 임의의 문자가 1개 이상 있으면), 1000won : 검색할 문자열
if  (lookahead1 != None):
    print(lookahead1.group())
else:
    print('None')
```

```python
lookahead2 = re.search(r'.+(?=am)', '2023-01-26 am 10:00:01')    # .+(?=am) => .+ : 'am' 문자열 앞에 1개 이상의 글자
print(lookahead2.group())
```

```python
# 전방 부정 탐색 (?!)
lookahead3 = re.search(r'\d{4}(?!-)', '010-1234-5678')    # 4자리 정수가 앞에 위치하고 앞에 '-'기호가 없을 때
print(lookahead3.group())
```

##### 후방 탐색(lookbehind)

```python
# 후방 긍정 탐색
lookbehind1 = re.search(r'(?<=am).+', '2023-01-26 am 11:10:01')    # am : 패턴
print(lookbehind1)

lookbehind2 = re.search(r'(?<=:).+', 'USD: %51')    # ':' : 패턴
print(lookbehind2)
```

```python
# 후방 부정 탐색
lookbehind3 = re.search(r'\b(?<!\$)\d+\b', 'I paid $30 for 100 apples.')
print(lookbehind3)    # \b : 공백, ?<! : 후방부정, \$ : 숫자, \b : 공백
# => 공백 뒤에 $ 기호가 없고 숫자 1개 이상 그리고 공백
```
