#### 1. OTT 선호도 조사 및 pie chart로 표시하세요.

```python
import matplotlib.pyplot as plt
import numpy as np
import koreanize_matplotlib
import random

ott_company = ['Netflix', 'Amazon', 'Disney+', 'Apple TV', 'TVing']
ott_dict = dict.fromkeys(ott_company, 0)    # 빈 딕셔너리 생성
#print(ott_dict)

for i in range(5000):
    ott_dict[random.choice(ott_company)] += 1
    # 랜덤으로 선택된 ott_company의 키값이 ott_dict에 올 때 마다 값 1 증가

#print(ott_dict)
vote = list(ott_dict.values())
#print(vote)


plt.pie(vote, labels=ott_company, startangle=90, autopct='%.1f%%')
plt.title('OTT 선호도 조사')
plt.show()
```

#### 2. 2020년 건강 검진 데이터를 이용한 histogram, boxplot, scatter 그리기

```python
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib

df = pd.read_excel('health_2020_1000ea.xlsx')

df_men = df[df['gender'] == 1]
#print(df_men)
height_men = list(df_men['height'])
#print(height_men)

df_women = df[df['gender'] == 2]
#print(df_women)
height_women = list(df_women['height'])
print(height_women)

plt.hist(height_men, bins=7, label=['남성'], alpha=0.5)
plt.hist(height_women, bins=7, label=['여성'], alpha=0.5)
#plt.hist([height_men, height_women], bins=7, label=['남성','여성'], alpha=0.5)
#이거는 남녀 키가 달라 범위가 달라져버려서 이렇게 나오는 것
plt.xlabel('키 분포')
plt.ylabel('횟수')
plt.title('2020년 건강 검진 데이터 남성 여성의 키 분포')
plt.grid(True, alpha=0.7)
plt.xlim([130, 195])
plt.legend()
plt.show()
```

```python
weight_men = list(df_men['weight'])
weight_women = list(df_women['weight'])

plt.boxplot([weight_men, weight_women], notch=True, orientation='vertical')

plt.xticks([1,2], ['남성', '여성'])
plt.xlabel('성별')
plt.ylabel('몸무게')
plt.title('2020년 건강 검진 데이터 남성 여성의 몸무게 분포')
plt.show()
```

```python
cholesterol = df['cholesterol']
blood_sugar = df['blood_sugar']

colors = df['drinking'].map({0:'blue', 1:'red'})
labels = df['drinking'].map({0: '비음주', 1:'음주'})

plt.scatter(cholesterol, blood_sugar, c=colors, edgecolor='k', alpha=0.5)
plt.xlabel('콜레스테롤')
plt.ylabel('혈당')
plt.title('콜레스테롤 vs. 혈당 관계')
plt.show()
```

#### 3. 2020년 건강 검진 데이터를 이용한 3개의 histogram 그리기(subplot)

```python
all_height = df['height']
under_160 = all_height[df['height'] <= 160]
over_170 = all_height[df['height'] >= 170]

fig, axes = plt.subplots(1,3,figsize=(9, 4))

axes[0].hist(all_height, bins=10, color='skyblue', edgecolor='k')
axes[0].set_title('All Heights')
axes[0].set_xlabel('Height (cm)')
axes[0].set_ylabel('Frequency')

axes[1].hist(under_160, bins=6, color='g', edgecolor='k')
axes[1].set_title('Height <= 160')
axes[1].set_xlabel('Height (cm)')
axes[1].set_ylabel('Frequency')


axes[2].hist(over_170, bins=6, color='orange', edgecolor='k')
axes[2].set_title('Height >= 170')
axes[2].set_xlabel('Height (cm)')
axes[2].set_ylabel('Frequency')


plt.suptitle('남여 키 분포 분석')
plt.tight_layout()
plt.show()
```
