### 부채꼴 스타일 지정

```python
import matplotlib.pyplot as plt

# wedgeprops : 딕셔너리 형태, width는 0~1 사이 값(전체 반지름에 대한 쐐기의 두께 비율)
ratio = [34,32,16,18]
labels=['Apple','Banana','Melon','Grapes']
colors=['#ff9999', '#ffc000', '#8fd9b6', '#d395d0']

wedgeprops={'width':0.7, 'edgecolor':'w', 'linewidth':5}
# 'edgecolor':'w', 'linewidth':5 => 각 값 사이에 흰색으로 두께 5인 칸 생성

plt.pie(ratio, labels=labels, autopct='%.1f%%',
        startangle=260, counterclock=False, colors=colors,
        wedgeprops=wedgeprops)    # counterclock=False : 시계방향

plt.show()
```

### 누적 히스토그램(Cumulative Histogram)

```python
import matplotlib.pyplot as plt
import numpy as np

# 정규분포 데이터 생성 loc:평균, scale:표준편차, size:데이터 개수
data1 = np.random.normal(loc=0, scale=1, size=1000)
data2 = np.random.normal(loc=1, scale=1.5, size=1000)

# 누적 히스토그램 그리기
plt.hist([data1, data2], bins=30, cumulative=True, label=['Data 1', 'Data 2'], color=['skyblue', 'salmon'], edgecolor='k', alpha=0.7)
# cumulative=True 로 누적 히스토그램, 빠지면 정규분포 그래프 그려짐

plt.title('Cumulative Histogram Example')
plt.xlabel('Value')
plt.ylabel('Cumulative Frequency')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

### 박스 플롯 기본 예제

```python
import matplotlib.pylab as plt
import numpy as np

# 랜덤 데이터 생성(세 그룹)
np.random.seed(0)
data1 = np.random.normal(50,10,100)    # 중간(10)은 표준 편차
data2 = np.random.normal(60,15,100)    # 중간(15)은 표준 편차
data3 = np.random.normal(55,20,100)    # 중간(20)은 표준 편차
# data3 처럼 표준편차가 커질수록 이상치(극단값) 생김

data=[data1, data2, data3]

# 박스플롯 그리기
plt.boxplot(data, orientation='vertical')

# x축 그룹 이름 지정
plt.xticks([1,2,3], ['Group 1', 'Group 2', 'Group 3'])
plt.title('Simple Boxplot Example')
plt.ylabel("Values")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

### 기본 바이올린 플롯 예제

### 바이올린 플롯 : 색상, 스타일 지정 예제

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(2)

data=[np.random.normal(m,1,200) for m in [0,1,2,3]]    # 한번에 4개의 데이터셋이 만들어짐

fig, ax = plt.subplots(figsize=(8,4))

# violins = ax.violinplot() : 바이올린 플롯의 각 구성 요소에 해당하는 목록들을 딕셔너리 형태로 반환
violins = ax.violinplot(data,showmeans=False, 
                        showmedians=True,
                        showextrema=True)
# showmeans : 평균값 표시, showmedians : 중앙값 표시, showextrema : 최대/최소값 표시

colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']

for i, body in enumerate(violins['bodies']):    # bodies 4개의 객체 중 하나씩 body에 전달
    body.set_facecolor(colors[i])    # 색상 값 채우기
    body.set_edgecolor('k')
    body.set_alpha(0.7)

# 분포의 중심선 색상 지정
violins['cbars'].set_edgecolor('k')

# 분포의 최대값 색상 지정
violins['cmaxes'].set_edgecolor('darkgray')

# 분포의 최소값 색상 지정
violins['cmins'].set_edgecolor('darkgray')

# 분포의 중앙값 표시
violins['cmedians'].set_edgecolor('white')

ax.set_title('Customized Violin Plot')
ax.set_xlabel('Group')
ax.set_ylabel('Value')
ax.set_xticks([1,2,3,4])
ax.set_xticklabels(['G1', 'G2', 'G3', 'G4'])
ax.grid(axis='y', linestyle='--', alpha=0.4)
plt.show()
```

### 여러 개 그래프 그리기 : subplot() 예제

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,2*np.pi, 100)    # 100개 균등 구간으로 나눔
y1=np.sin(x)
y2=np.cos(x)
y3=np.tan(x)

# 위 셀에서 두께가 두꺼워졌으니 기본값인 1.5로 바꿔줌
plt.rcParams['lines.linewidth']=1.5
# 이렇게 한번에 기본값으로 되돌리기 가능
# rcParams은 개별, style은 그 묶음 개념으로 이해
plt.style.use('default')

plt.subplot(3,1,1)    # 3행 1열 중 첫 번째(맨 위)
plt.plot(x, y1)
plt.title('y=sin(x)')

plt.subplot(3,1,2)    # 3행 1열 중 두 번째(중간)
plt.plot(x, y2)
plt.title('y=cos(x)')

plt.subplot(3,1,3)    # 3행 1열 중 세 번째(맨 아래)
plt.plot(x, y3)
plt.title('y=tan(x)')

plt.tight_layout()
plt.show()
```

### 여러 개 그래프 그리기 subplot(): 축 공유하기

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,2*np.pi,100)
y1 = np.sin(x)
y2 = np.cos(x)

ax1=plt.subplot(2,1,1)
plt.plot(x,y1)
plt.title('y=sin(x)')
plt.xticks(visible=False)

# 축 눈금 공유 -> sharex(x축 눈금 공유), sharey(y축 눈금 공유)
plt.subplot(2,1,2,sharex=ax1)
plt.plot(x,y2)
plt.title('y=cos(x)')

plt.tight_layout()
plt.show()
```

### subplots(2,2) 예제

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.sin(x+1)
y3=np.sin(x+2)
y4=np.sin(x+3)

# 2행 2열 subplot 생성, x축과 y축 모두 공유
fig, axes=plt.subplots(2,2, sharex=True, sharey=True, figsize=(8,6))
fig.suptitle('This is the Figure Title', fontsize=15)

# axes[행(row), 열(col)] 이렇게 2차원 배열처럼 접근, 이는 numpy방식이다. 파이썬은 다름, 메모장 19일차 참조
axes[0,0].plot(x,y1)
axes[0,0].set_title('Plot 1')

axes[0,1].plot(x,y2)
axes[0,1].set_title('Plot 2')

axes[1,0].plot(x,y3)
axes[1,0].set_title('Plot 3')

axes[1,1].plot(x,y4)
axes[1,1].set_title('Plot 4')

plt.tight_layout()
plt.show()
```

### 두 종류의 그래프 그리기 : plot과 bar 차트

```python
import matplotlib.pyplot as plt
import koreanize_matplotlib

# 1. 데이터 생성
months= ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
temperature=[2,4,8,15,20,24]    # 단위: 섭씨(℃)
precipitation=[70,60,50,40,30,20]    # 단위: 밀리미터(mm)

# 2. 기본 그래프 설정
fig, ax1=plt.subplots(figsize=(8,5))

# 3. 첫 번째 Y축 (왼쪽) - 기온
color1 = 'orangered'
ax1.set_xlabel('Month')
ax1.set_ylabel('Temperature (℃)')
ax1.plot(months, temperature, color=color1, marker='o', linestyle='-')

# Y축 레이블 색상 생성
#ax1.tick_params(axis='y', labelcolor=color1)

# 4. 두 번째 Y축 (오른쪽) 생성을 위해 X축 공유
ax2=ax1.twinx()    # 새로운 Axes(ax2) 생성

# 5. 두 번째 Y 축 (오른쪽) - 강수량
color2='royalblue'
ax2.set_ylabel('Precipitation (mm)', color='b')
ax2.bar(months, precipitation, color=color2, alpha=0.5)

# Y축 레이블 색상 생성
#ax2.tick_params(axis='y', labelcolor=color2)

# 6. 추가 설정
plt.title('온도와 강수량 그래프')
plt.show()
```

### 그래프에 텍스트 삽입 예제 2

```python
import matplotlib.pyplot as plt
import pandas as pd

data={'Name':['Alex', 'Bob', 'Clarein', 'Dexter'],
      'Score': [45,23,78,65]}
df=pd.DataFrame(data)

plt.figure(figsize=(8,6))
bars=plt.bar(df['Name'], df['Score'], color='orange')

# Annotating bars
for bar in bars:    # 4개의 데이터 중 하나씩 가져옴
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(),
             f'{bar.get_height():.2f}',    # 각 bar chart의 높이를 소수점 둘째 자리까지 출력
             ha='center', va='bottom',fontsize=12)    # ha : 수평정렬, va : 수직정렬
    # get_x() + bar.get_width()/2 : x 좌표, 그래프의 왼쪽 끝 값이 bar.get_x()이고 여기에 그래프 폭의 절반인 bar.get_width()/2을 더한 것
    # bar.get_height() : y 좌표

# Adding labels and title
plt.xlabel('Name', fontsize=14)
plt.ylabel('Score', fontsize=14)
plt.title('Final Exam Score', fontsize=16)
plt.show()
```

### 그래프에 주석 추가 예제
- annotate() 함수

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,10,100)
y=np.sin(x)

plt.plot(x,y,label='sin(x)')
plt.plot(np.pi/2, 1, 'o', color='k')

# annotate(text, xy, xytext=None, xycoords=‘data’, textcoords=None,arrowprops=None, ...)
# text : 표시할 주석 텍스트(문자열)
# xy : 주석을 붙일 데이터 좌표(x,y) - 화살표가 가리키는 위치
# xytext : 주석 텍스트의 위치 좌표(x,y)
plt.annotate('Peak point', xy=(np.pi/2,1), xytext=(4,0.75), 
             arrowprops=dict(facecolor='r', shrink=0.2),
             fontsize=12, color='r')

plt.title("Sine Wave with Annotation")
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
plt.grid(True)
plt.show()
```
