### 그래프 옵션 : 범례(legend) 추가

```python
import matplotlib.pyplot as plt
DASHED_LINE = '--'

x_axis = [0,1,2,3]    # x축 tick
y_axis = [10,20,30,40,50]    # y축 tick

data1 = [10,20,30,40]
data2 = [20,30,40,50]

plt.title('Title & Legend Test')
plt.plot(x_axis, data1, marker = 'o', label='data1', linestyle=DASHED_LINE)
plt.plot(x_axis, data2, marker = 's', label='data2')    # label 값 : legend에 표시할 이름
# 2번의 plot 호출

plt.legend(loc='lower center', ncols=2)

plt.xticks(x_axis)
plt.yticks(y_axis)
plt.show()
```

### 그래프 영역 채우기 : fill_between()

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [2,3,5,10]

plt.plot(x,y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.ylim(0,10)
plt.xlim(1,4)
plt.fill_between(x[1:3], y[1:3], color='lime', alpha=0.5)    # alpha : 투명도
plt.show()
```

### 두 그래프 사이 영역 채우기 : fill_between()

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y1=[2,3,5,10]
y2=[1,2,4,8]

plt.plot(x,y1)
plt.plot(x,y2)

plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.fill_between(x[1:3], y1[1:3], y2[1:3], color='skyblue', alpha=0.5)
plt.show()
```

### 눈금 표시하기

### 가로 막대 그래프(barh)

```python
import matplotlib.pyplot as plt

# 데이터 준비
categories = ['A','B','C','D']
values = [5,7,3,8]

# 가로 막대 차트 그리기
plt.barh(categories, values, color='skyblue',
         edgecolor='k', height=0.5, alpha=0.7)

# 제목과 축 레이블 추기
plt.title('Horizontal Bar Chart Example')
plt.xlabel('Values')
plt.ylabel('Categories')

# 그래프 출력
plt.show()
```

### 여러 항목의 데이터를 가지는 막대 그래프

```python
import matplotlib.pyplot as plt
import numpy as np

labels= ['A','B','C']    # x축 카테고리
values_2023 = [20,34,30]    # 첫 번째 항목 데이터
values_2024 = [25,32,34]    # 두 번째 항목 데이터

x = np.arange(len(labels))    # 카테고리 위치(0,1,2,...) => x = [0,1,2,...]
width = 0.4    # 각 막대의 너비(default 너비값이 0.8이므로 잘 겹치기 위해 위치 조정한 것)

plt.bar(x-width/2, values_2023, width, label='2023', color='royalblue')
plt.bar(x+width/2, values_2024, width, label='2024', color='salmon')

# 축/제목/눈금/범례 설정
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Grouped Bar Chart Example')
plt.xticks(x, labels)
plt.legend()
plt.show()
```

### 세로 누적 막대 그래프 : bar(bottom 속성)

```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['A','B','C','D']
data1 = np.array([10,20,10,30])    # numpy의 array 사용
data2 = np.array([20,25,15,25])    # 배열의 누적 연산을 쉽게 하기 위함
data3 = np.array([12,15,19,6])    # array : 배열, list : 가변길이

# 첫 번째 막대 그리기
plt.bar(categories, data1, bottom=0, color='royalblue', label='Data 1')

# 두 번째 막대 그리기(data1 위에 누적)
plt.bar(categories, data2, bottom=data1, color='gold',label='Data 2')

# 세 번째 막대 그리기(data1 + data2 위에 누적)
plt.bar(categories, data3, bottom=data1+data2, color='salmon', label='Data 3')
# bottom=data1+data2 : 각 배열의 인덱스 별로 덧셈 연산

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Stacked Bar Chart Example')
plt.legend()    # 범례 표시
plt.show()
```

### 가로 누적 막대 그래프 예제 : barh(left 속성)

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 준비
categories = ['Xiaomi', 'Apple','Samsung']
group1 = np.array([5,10,15])    # 2023년
group2 = np.array([3,6,9])    # 2024년
group3 = np.array([2,4,6])    # 2025년

# 누적 가로 막대 그리기
plt.barh(categories, group1, left=0, color='orangered', label='2023')
plt.barh(categories, group2, left=group1, color='orange', label='2024')
plt.barh(categories, group3, left=group1+group2, color='limegreen', label='2025')

plt.title('Stacked Horizontal Bar Chart Example')
plt.xlabel('Value')
plt.legend()
plt.show()
```

### DataFrame을 이용한 누적 막대 그래프 그리기 : 방법 1

```python
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib

# 1. DataFrame 생성
data = {'도시':['서울', '부산', '인천', '대전'],
        '사과':[30,20,15,25],
        '바나나':[10,25,20,30],
        '오렌지':[20,10,25,15]}

df = pd.DataFrame(data)
df = df.set_index('도시')    # 도시 열을 인덱스로 설정

print(df)

value1 = df['사과']    # df['사과'].to_numpy() 또는 df['사과'].values 이렇게도 가능
value2 = df['바나나']
value3 = df['오렌지']

# df.index 대신 df.inded.to_list() 사용
plt.bar(df.index, value1, bottom=0, color='#FF6B6B', label='사과')
plt.bar(df.index, value2, bottom=value1, color='#4ECDC4', label='바나나')
plt.bar(df.index, value3, bottom=value1+value2, color='#CFD145', label='오렌지')

plt.xlabel('도시')
plt.ylabel('판매량')
plt.legend(title='과일 종류', loc='upper left')    # 범례 위치 조정
plt.xticks(rotation=0)    # x축 레이블 회전 방지
plt.grid(axis='y', linestyle='--', alpha=0.6)    # y축 그리드 라인 추가

plt.show()
```

### DataFrame을 이용한 누적 막대 그래프 그리기 : 방법 2

```python
# DataFrame.plot(kind='bar', stacked=True)사용
# stacked=True를 사용하면 매 bottom 마다 입력해줘야 하는 수고 덜 수 있음
# 누적 막대 그래프 그리기
df.plot(kind='bar', stacked=True, figsize=(10,6), color=['#FF6B6B', '#4ECDC4', '#CFD145'])
# 앞 shell의 값을 어떻게 가져왔느냐? = df

# 그래프 꾸미기(제목, 축 레이블 등)
plt.title('도시별 과일 판매량 누적 막대 그래프')
plt.xlabel('도시')
plt.ylabel('판매량')

plt.legend(title='과일 종류', bbox_to_anchor=(1.01, 1))    # bbox_to_anchor : 범례를 고정할 못(Anchor)의 위치
# bbox_to_anchor=(1.01, 1) 왜 1.01인가? 그래프 테두리 끝 좌표가 (1.0, 1.0)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.show()
```
