#### 1. 연도별 매출 데이터를 이용한 선 그래프 및 막대 그래프 그리기

```python
import matplotlib.pyplot as plt
import koreanize_matplotlib

years = [2020, 2021, 2022, 2023, 2024, 2025]
data = [150, 180, 160, 200, 250, 260]

plt.plot(years, data, marker='o', color='b')
plt.title('매출액 선 그래프(2020-2025)')
plt.xlabel('연도')
plt.ylabel('매출액')
plt.grid(True, axis='both', color='gray', alpha=0.5, linestyle='--')
plt.show()
```

```python
plt.bar(years, data, color='orange')
plt.title('매출액 막대 그래프(2020-2025)')
plt.xlabel('연도')
plt.ylabel('매출액')
plt.grid(True, axis='y', color='gray', alpha=0.5, linestyle='--')
plt.show()
```

#### 2. 6개월 동안 방문객 및 수익 비교

```python
import matplotlib.pyplot as plt

months = [1,2,3,4,5,6]
visitors = [120, 135, 160, 180, 200, 220]
revenue = [240, 270, 320, 360, 400, 430]

plt.plot(months, revenue, marker='o', color='r', label='Revenue')
plt.bar(months, visitors, color='skyblue', label='Visitors')
plt.title('Monthly Visitors and Revenue')
plt.xlabel('Month')
plt.ylabel('Count/Revenue')
plt.grid(True, axis='y', color='gray', alpha=0.5, linestyle='--')
plt.legend(loc='upper left')
plt.show()
```

#### 3. 분기별 판매량 누적 막대 그래프

```python
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib

datas = {'Quarter': ['Q1', 'Q2', 'Q3','Q4'],
         'Product_A': [20,35,30,35],
         'Product_B': [25,32,34,20],
         'Product_C': [15,18,22,28]}

df = pd.DataFrame(datas)
df = df.set_index('Quarter')    # Quarter 열을 인덱스로 지정

df.plot(kind='bar', stacked=True, color=['royalblue', 'darkorange', 'green'])

plt.title('분기별 판매량 누적 막대 그래프')
plt.xlabel('분기')
plt.ylabel('판매량')

plt.legend(title='판매량', fontsize=8)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```
